###Usando PCA no dataset da NBA

```python
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA

df = pd.read_csv('NBA_stats.csv')

df.head()
```

```python
features = list(df.columns[5:-2]) #colunas 5 até antepenúltima são contínuas e numéricas
```

Normalizando e fazendo PCA:

```python
df_norm = df[features]
df_norm=(df_norm-df_norm.min())/(df_norm.max()-df_norm.min())


pca = PCA(n_components=2)
components = pca.fit_transform(df_norm[features])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(2)
    )
fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
display(pd.DataFrame(pca.components_, columns = features, index = ['PC 1', 'PC 2']))
```

Da tabela acima parece que a quantidade de minutos jogados e a quantidade de tentativas de lances de 3 pontos (!!!) são as features com maior variância e explicabilidade do dataset.
